# **`IMPORTAMOS LIBRERIAS`**

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

# **GENERACIÓN DIMENSION FECHA**

```python
start_date = datetime(2022,1,1)
end_date = datetime(2024,12,31)

dates = pd.date_range(start_date, end_date)

dim_date = pd.DataFrame({
    "full_date": dates
})

dim_date["date_id"] = dim_date["full_date"].dt.strftime("%Y%m%d").astype(int)
dim_date["day"] = dim_date["full_date"].dt.day
dim_date["month"] = dim_date["full_date"].dt.month
dim_date["month_name"] = dim_date["full_date"].dt.strftime("%B")
dim_date["quarter"] = dim_date["full_date"].dt.quarter
dim_date["year"] = dim_date["full_date"].dt.year
dim_date["week_day"] = dim_date["full_date"].dt.strftime("%A")

dim_date = dim_date[[
    "date_id",
    "full_date",
    "day",
    "month",
    "month_name",
    "quarter",
    "year",
    "week_day"
]]

dim_date.head()
```

# **GENERACIÓN DIMENSIÓN CLIENTES**

```python
#Creamos las listas

first_names_male = ["Juan","Carlos","Luis","Pedro","Andrés",
                    "Miguel","Jorge","David","Daniel","Fernando"]
first_names_female = ["María","Ana","Laura","Sofía","Paula","Camila",
                      "Valentina","Andrea","Carolina","Diana"]

last_names = ["Gómez","Rodríguez","Martínez","López","Hernández",
              "García","Pérez","Sánchez","Ramírez","Torres"]

cities = [
    "Bogotá",
    "Medellín",
    "Cali",
    "Barranquilla",
    "Cartagena",
    "Bucaramanga",
    "Pereira",
    "Manizales",
    "Santa Marta",
    "Ibagué"
]

segments = ["Regular","Premium","VIP"]
```

```python
#Generamos los clientes

num_customers = 1500

customers = []

start_date = datetime(2020,1,1)
end_date = datetime(2024,12,31)

for i in range(num_customers):

    gender = random.choice(["Male","Female"])

    if gender == "Male":
        name = random.choice(first_names_male)
    else:
        name = random.choice(first_names_female)

    last = random.choice(last_names)

    full_name = name + " " + last

    age = random.randint(18,65)

    city = random.choice(cities)

    registration_date = start_date + timedelta(
        days=random.randint(0,(end_date-start_date).days)
    )

    segment = random.choices(
        segments,
        weights=[0.7,0.2,0.1]
    )[0]

    customers.append([
        full_name,
        gender,
        age,
        city,
        registration_date,
        segment
    ])
```

```python
#Creamos el DataFrame

dim_customers = pd.DataFrame(customers, columns=[
    "full_name",
    "gender",
    "age",
    "city",
    "registration_date",
    "customer_segment"
])

dim_customers.head()
```

# **GENERAMOS FACT_SALES**

```python
# Cargar los date_id reales
dates = pd.read_csv("dim_date_ids.csv")
date_ids = dates["DATE_ID"].tolist()

products = pd.read_csv("dim_products.csv")

product_ids = products.index + 1
product_ids = product_ids.tolist()

fact_sales = []

num_sales = 50000

for i in range(num_sales):

    date_id = random.choice(date_ids)

    customer_id = random.randint(1,1500)

    product_id = random.choice(product_ids)

    quantity = random.randint(1,5)

    unit_price = round(random.uniform(20,2000),2)

    total_amount = round(quantity * unit_price,2)

    fact_sales.append([
        date_id,
        customer_id,
        product_id,
        quantity,
        unit_price,
        total_amount
    ])
```

```python
#Creamos el DataFrame
fact_sales = pd.DataFrame(fact_sales, columns=[
    "date_id",
    "customer_id",
    "product_id",
    "quantity",
    "unit_price",
    "total_amount"
])
print(fact_sales)
```
